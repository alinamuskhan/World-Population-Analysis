```python
df['World Population Percentage'] = df['2022 Population'] - df['2015 Population']
```

```python
df['World Population Percentage'].value_counts().head()
```

```python
dg= df.sort_values(by='World Population Percentage', ascending=False).head()   
dd= df.sort_values(by='World Population Percentage', ascending=True).head() 
```

```python
print("Countries with the most population growth from 2015 to 2022:")
print()
print(dg[['Country','World Population Percentage']])

print("\nCountries with the most population decline from 2015 to 2022:")
print()
print(dd[['Country','World Population Percentage']])
```
